```python
# Definir una función que aplica estilos a columnas específicas con colores diferentes
def resaltar_columnas(s, colores):
    return ['background-color: {}'.format(colores.get(col, '')) for col in s.index]
```

#### Filtrado datos de Dr. Christian Moni Bidin (cmoni@ucn.cl)

Filtramos aquellos que tienen 3 o más columnas consecutivas.

Las columnas resaltadas en colores representan días consecutivos.

```python

# Leer el archivo Excel
df = pd.read_excel("List_Obs_in_Data.xlsx")

# Filtrar el DataFrame por el "Principal Investigator (PI)"
df_christian = df[df["Principal Investigator (PI)"] == "Dr. Christian Moni Bidin (cmoni@ucn.cl)"]

# Convertir todas las columnas relevantes a tipo numérico (ignorar errores)
# Especificamos columns=slice(start, end) para seleccionar solo las columnas que queremos convertir
df_christian.iloc[:, 8:] = df_christian.iloc[:, 8:].apply(pd.to_numeric, errors='coerce')

# Función para verificar si hay 5 valores consecutivos diferentes de cero en una fila
def fechas_consecutivas(row):
    count = 0
    for value in row:
        if value != 0 and not pd.isna(value):
            count += 1
            if count >= 3:
                return True
        else:
            count = 0
    return False

# Aplicar la función a cada fila del DataFrame
df_christian_filtered = df_christian[df_christian.iloc[:, 8:].apply(fechas_consecutivas, axis=1)]


# Seleccionar columnas que tienen al menos un valor diferente de cero
pd.set_option('display.max_columns', None)
non_zero_columns_christian = df_christian_filtered.loc[:, (df_christian_filtered != 0).any(axis=0)]

# Especificar los colores para cada columna
colores = ["#D4C685","#F7EF81","#CFE795","#A7D3A6","#ADD2C2"]
colores_resaltar = {datetime.datetime(2022, 4, 21, 0, 0): colores[0],datetime.datetime(2022, 4, 22, 0, 0): colores[0],datetime.datetime(2022, 4, 23, 0, 0): colores[0],datetime.datetime(2023, 3, 30, 0, 0): colores[1],datetime.datetime(2023, 3, 31, 0, 0): colores[1],datetime.datetime(2023, 7, 27, 0, 0): colores[2],datetime.datetime(2023, 7, 28, 0, 0): colores[2]}

# Aplicar estilos al DataFrame
estilos = non_zero_columns_christian.style.apply(lambda x: resaltar_columnas(x, colores_resaltar), axis=1)

estilos
```

```python
# Obtén la ruta absoluta
file_path = os.path.abspath('Nrich/news_RV/Objetos_UCN.csv')
print(file_path)

# Luego intenta leer el archivo
df_UCN = pd.read_csv(file_path)
df_UCN
```

```python
# Convertir la columna de fechas a tipo datetime
df_UCN['Fecha'] = pd.to_datetime(df_UCN['Fecha'])

# Agrupar por objeto y obtener las fechas en las que se encuentran
grouped = df_UCN.groupby("Objetos")['Fecha'].apply(list).reset_index()

# Mostrar el resultado
grouped
```

```python
conteo_nombres = df_UCN["Objetos"].value_counts()
conteo_nombres
```

```python
# Agrupar por la columna 'Objetos' y sumar los valores de la columna 'Rep'
suma_reps_por_objeto = df_UCN.groupby("Objetos")["Rep"].sum()

# Mostrar el resultado
df_conteo = suma_reps_por_objeto.reset_index()

# Renombrar las columnas para mayor claridad si es necesario
df_conteo.columns = ["Objetos", 'Suma_Rep']
df_conteo = df_conteo.sort_values(by='Suma_Rep', ascending=False)
df_conteo = df_conteo[df_conteo["Suma_Rep"] != 1]
df_conteo
```

```python
df_conteo.to_csv('probables_estándar.csv', index=False)
```
